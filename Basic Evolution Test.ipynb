```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def distance(X, Y):
    return np.sqrt(np.sum(np.square(X-Y),axis=1))
```

```python
def geneticMutation(gene, geneMutation=0.2, noMutations=20, mutationIntensity=1.0):
    Gene = np.stack([gene]*noMutations)
    bino = np.random.binomial(1,geneMutation, size=Gene.shape)
    bino = bino*mutationIntensity
    muta = np.random.randn(*Gene.shape)
    Gene = Gene + bino*muta
    return Gene
```

```python
nGenerations = 2000
generationPopulation = 20
geneticCode = 200
geneMutation = 0.01
mutationIntensity = 0.1

population = np.random.randn(generationPopulation, geneticCode)
target = np.random.randn(1, geneticCode)
genBestD = []
genWorstD = []
for generation in range(nGenerations):
    D = distance(population, target)
    print(f"Gen: {generation} and Distance: {D.min()}")
    population = geneticMutation(population[D.argmin()], geneMutation, generationPopulation, mutationIntensity)
    genBestD.append(D.min())
    genWorstD.append(D.max())
fittestGene = population[D.argmin()]
```

```python
plt.plot(genBestD,label='Best')
plt.plot(genWorstD, label='Worst')
plt.legend()
```
