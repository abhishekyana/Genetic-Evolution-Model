```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def distance(X, Y):
    """
    Euclidean Distance between X and Y.
    """
    return np.sqrt(np.sum(np.square(X-Y),axis=1))
```

```python
def geneticMutation(gene, geneMutation=0.02, noMutations=20, mutationIntensity=1.0):
    """
    This function takes a gene and gives out mutated genes or offsprings for next generation.
    Args:
    gene: Fittest Gene of the previous generation.
    geneMutation: percentage of genecode to be altered or mutated. Default: 2%
    noMutations: # of mutations of offsprings required for next generation. Default: 20
    mutationIntensity: This factor value to will decide how strong the mutation should be. Default: 1.0
    
    Outputs:
    Genes: Mutated genes.
    """
    Genes = np.stack([gene]*noMutations)
    bino = np.random.binomial(1,geneMutation, size=Genes.shape)
    bino = bino*mutationIntensity
    muta = np.random.randn(*Genes.shape)
    Genes = Genes + bino*muta
    return Genes
```

```python
nGenerations = 3000
generationPopulation = 20
geneticCode = 200
geneMutation = 0.01
mutationIntensity = 0.1

population = np.random.randn(generationPopulation, geneticCode)
target = np.random.randn(1, geneticCode)
target = np.sin(np.linspace(-10,10,200)).reshape(1,-1)
genBestD = []
genWorstD = []
somePlts = []
for generation in range(nGenerations):
    D = distance(population, target)
    if generation%200==0:
        somePlts.append(population[D.argmin()])
    print(f"Gen: {generation} Distance: {D.min()}")
    population = geneticMutation(population[D.argmin()], geneMutation, generationPopulation, mutationIntensity)
    genBestD.append(D.min())
    genWorstD.append(D.max())
fittestGene = population[D.argmin()]
```

```python
plt.plot(genBestD,label='Best')
plt.plot(genWorstD, label='Worst')
plt.legend()
```
